```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

```python
fig = plt.figure(figsize=(16, 9), dpi=100)
ax = fig.add_subplot(1, 2, 1)

xs: np.ndarray = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
ys: np.ndarray = np.sin(xs)

ax.plot(xs, ys, 'D-k', label="sin(x)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Graph powered by Matplotlib")
ax.legend()

ax2 = fig.add_subplot(1, 2, 2)
data = pd.Series(list(range(10)))
data.index = [chr(i + ord('a')) for i in range(10)]

ax2.plot(data, 'o-k', label='hoge')
ax2.vlines(0.5, 0, 10)
ax2.hlines(5, 0, 10)
ax2.legend()

fig.legend()

print(ax2.xaxis.get_major_formatter())
```

```python
xs: np.ndarray = np.random.randn(100)
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)

n, bins, patches = ax.hist(xs, bins="fd", edgecolor="black", linewidth=1.2, range=(-10, 10))
ax.set_title("Histogram powered by Matplotlib", fontsize=24, pad=20)
# ax.set_xticks(bins)
# ax.set_yticks(np.arange(0, float(max(n) + 1), 2))
ax.xaxis.set_major_locator(ticker.FixedLocator(bins))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.grid(visible=True, which="major", linewidth=0.5)
ax.tick_params(axis="x", rotation=45)
```
